```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
dataset.head()
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

Training the Random Forest Regression on the entire dataset



```python
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor(n_estimators=10,random_state=0)
regressor_rf.fit(X,y)
```

Predicting the new result

```python
regressor_rf.predict([[6.5]])
```

Visualising the Random Forest Regression result(higher resolution)

```python
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color = 'red')
plt.plot(X_grid,regressor_rf.predict(X_grid),color='blue')
plt.title('Truth or Bluff Random Forest Regression model)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
