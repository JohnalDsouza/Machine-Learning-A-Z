```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
# check if there is a need of applying data preprocessing tools
# like check missing data or encoding categorical data
# NO FEATURE SCALING
```

Training the Decision Tree Regression Model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
regression_tree = DecisionTreeRegressor(random_state=0)
regression_tree.fit(X, y)
```

Predicting the result

```python
regression_tree.predict([[6.5]])
```

Visualising the DTR results in higher resolution

```python
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color = 'red')
plt.plot(X_grid,regression_tree.predict(X_grid),color='blue')
plt.title('Truth or Bluff Decision Tree Regression model)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
